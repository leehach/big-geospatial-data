### Control of Flow

#### Conditionals and `if` blocks

Python supports the usual logical conditions from mathematics:

Equals: `a == b`

Not Equals: `a != b`

Less than: `a < b`

Less than or equal to: `a <= b`

Greater than: `a > b`

Greater than or equal to: `a >= b`

The `if` block can use the `else` statement to do something if the main condition isn't passed.

```python
a = 33
b = 200

if b > a:
    print("b is greater than a")
else:
    print("b is smaller or equal to a")
```

An `if` block can have one or more `elif` statements to test alternative conditions. The `else` statement runs if no test is passed. The tests are evaluated *in order*, and the rest of the block is skipped over after the first `True` condition is encountered.

```python
a = 200
b = 33

if b > a:
    print("b is greater than a")
elif a == b:
    print("a and b are equal")
else:
    print("a is greater than b")
```

#### Looping with a `for` loop

A for loop is used for iterating over a collection. This can include core object types such as lists, tuples, and dictionaries (note, however, that dictionaries are unordered). It also includes strings, in which case each letter in the string is returned as an element.

This is less like the `for` keyword in other programming languages, and works more like an iterator method as found in other object-orientated programming languages. With the for loop we can execute a set of statements, once for each item in a list, tuple, set, etc.

```python
cities = ['San Francisco', 'Los Angeles', 'New York', 'Atlanta']

for city in cities:
    print(city)
```

```python
# define a list and then loop each item
list_ex = [1, '1', 1.0, 'one', [1], [1,2,3]]

for i in list_ex:
    print(i)
```

#### Looping with Conditions

```python
for city in cities:
    if city == 'Atlanta':
        print(city)
```

```python
for city in cities:
    if city == 'Atlanta':
        print(city)
    else:
        print('This is not Atlanta and is:', city)
```

```python
cities_population = {
    'San Francisco': 881549,
    'Los Angeles': 3792621,
    'New York': 8175133,
    'Atlanta':498044
}

for city, population in cities_population.items():
    
    if population < 1_000_000:
        size = "small"
    elif 1_000_000 <= population < 5_000_000:
        size = "big"
    else:
        size = "mega"
        
    print(f'{city} is a {size} city')
```

### Numpy Arrays

Why Use NumPy?

In Python we have lists that serve the purpose of arrays, but they are slow to process. NumPy aims to provide an array object that is up to 50x faster than traditional Python lists. The array object in NumPy is called ndarray, it provides a lot of supporting functions that make working with ndarray very easy. It is efficient way to manage arrays of numeric data 

```python
import numpy as np

a = np.array([1, 2, 3, 4])
print(a)
```

```python
arr = np.array([1, 2, 3, 4])

print('The first item in the array is:', arr[0])
print(arr[2] + arr[3])
```

Two dimensional array

### Open shapefile using Fiona

However, Geopandas is not able to handle really big data. Therefore, we can think about using **lower level** packages, which are more difficult to program but give you more flexibility. Here is an example of using `fiona`, which is a **dependency** module of `geopandas`. That is, rather than create its own I/O (import/export) routines, Geopandas uses the Fiona library for I/O, and offers a simpler syntax, but doesn't expose all of the capabilities of Fiona.

```python
# loop all features in the shapefile using fiona
import fiona
from shapely.geometry import shape

datasource = fiona.open(shpfile)
crs = datasource.crs

# loop each feature
for feat in datasource:

    # get the properties of each feature
    props = feat['properties'] # here feat is a dictionary
    
    # Then you can get the attributes of the feature using dictionary key-pair
    per_capita = props['per_capita']
    tract = props['tract']
    
    # get the geometry of the feature
    geom = feat['geometry']
    geom_obj = shape(geom)
    print('The per capita income of the tract', per_capita, tract)
    
```

### Open georeferenced Raster data using `rasterio`

`Geopandas` and `fiona` are good at handling vector data such as shapefiles. Using `rasterio` will allow you to open and analyze raster data. The following script will read raster data into a numpy array. Then you can manipulate and analyze the raster data.  

```python
import rasterio as rio

rasterfile = 'data/26928E233974N.tif'
raster = rio.open(rasterfile)

band = raster.read(1) #read the band 1, if you raster data is multi-bands, you can then get other bands
```

Check the dimension of the raster data

```python
band.shape
```

Then we can show the image of the raster

```python
plt.figure(figsize=(10, 10))
plt.imshow(band)
```
